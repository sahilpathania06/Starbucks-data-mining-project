# Recommending Data 

### A simple recommendation system is created in this notebook that helps in suggesting promotional offers to different customer segments.

## ---------------------------------------------------------------------------------

## 1. Importing Libraries

```python
import pandas as pd
import numpy as np

print("libraries are loaded successfully")
```

## -----------------------------------------------------------------------------------

## 2. Loading cleaned datasets (master and customers)

```python
master_data = pd.read_csv("../cleaned_data/master_data.csv")
print("Master file is loaded successfully")

customers = pd.read_csv("../cleaned_data/customers.csv")
print("Customer file is loaded successfully")
```

## ---------------------------------------------------------------------------------

## 3. Checking Available Offer Types

```python
# Here are all the types of offers

master_data["offer_type"].value_counts()
```

## ------------------------------------------------------------------------------

## 4. Most Common Offer Type

```python
master_data["offer_type"].mode()
```

## --------------------------------------------------------------------------------

## 5. Average Reward by Offer Type

```python
# It tells us that which offer gives us normally more reward

master_data.groupby("offer_type")["reward"].mean()
```

## ----------------------------------------------------------------------------------------

## 6. Creating Recommendations

```python
# Creating recommendations based on the age of the customers

def recommend_offer(segment):

    if segment == "Young":
        return 'bogo'
    
    elif segment == "Middle Age":
        return 'discount'
    
    else:
        return "informational"
```

## -------------------------------------------------------------------------

## 7. Generating Recommendations

```python
customers["recommend_offer"] = customers["age_segment"].apply(recommend_offer)

customers.head()
```
