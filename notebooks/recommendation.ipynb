# Recommending Data 

### A simple recommendation system is created in this notebook that helps in suggesting promotional offers to different customer segments.

## ---------------------------------------------------------------------------------

## 1. Importing Libraries

```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

pd.set_option("display.max_column", None)

print("libraries are loaded successfully")
```

## -----------------------------------------------------------------------------------

## 2. Loading cleaned Master Datest

```python
master_data = pd.read_csv("../cleaned_data/master_data.csv")
print("Master file is loaded successfully")
```

## ---------------------------------------------------------------------------------

## 3. Selecting Completed Offers

```python
# Here are all the types of offers

completed_offers = master_data[master_data["event"] == "offer completed"]

completed_offers.head()
```

## ------------------------------------------------------------------------------

## 4. Creating Customer Offer Matrix

```python
# Here rows are customers, columns are offers and values are how many times customer completed that offer

customer_offer_matrix = pd.crosstab(
    completed_offers["person"],
    completed_offers["offer_id"]
)

customer_offer_matrix.head()
```

## --------------------------------------------------------------------------------

## 5. Creating Offer Similarity Matrix

```python
# Using cosine similarity

offer_similarity = cosine_similarity(customer_offer_matrix.T)

offer_similarity_df = pd.DataFrame(
    offer_similarity,
    index = customer_offer_matrix.columns,
    columns = customer_offer_matrix.columns
)

offer_similarity_df.head()
```

## ----------------------------------------------------------------------------------------

## 6. Checking Similar Offers

```python
# This shows which of the the offers are most similar to the the selected offer

offer_id = customer_offer_matrix.columns[0]

offer_similarity_df[offer_id].sort_values(ascending=False).head()
```

## -------------------------------------------------------------------------

## 7. Recommendation Function

```python
def recommend_similar_offers(offer_id):
    similar_offers = offer_similarity_df[offer_id].sort_values(ascending=False)
    return similar_offers.head(5)
```

## -----------------------------------------------------------------------

## 8. Test Recommendation

```python
recommend_similar_offers(customer_offer_matrix.columns[0])
```

### Observation :-

### The selected offer has highest similarity with itself (1.0). The remaining offers show similarity scores around 0.22, this indicates that the customers who completed the selected offer also interacted with these offers. These offers can therefore be recommended as similar promotions.

## ---------------------------------------------------------------------------

## 9. Saving similarity matrix

```python
offer_similarity_df.to_csv("../cleaned_data/offer_similarity_matrix.csv")
```
