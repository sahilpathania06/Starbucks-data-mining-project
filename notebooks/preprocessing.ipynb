# Data Preprocessing

### In this notebook, we are gonna focus on cleaning and preparing the datasets of starbucks before integrating and analysing them 

## ---------------------------------------------------------------------------------

## 1. Import Libraries

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

print("The libraries are loaded successfully")
```

## -----------------------------------------------------------------------------------

## 2. Load Datasets

```python
profile = pd.read_csv("../data/profile.csv")
portfolio = pd.read_csv("../data/portfolio.csv")
transcript = pd.read_csv("../data/transcript.csv")

print("The datasets are loaded succusessfully")
```

## --------------------------------------------------------------------------------

## 3. Inspecting columnns of the Dataset

```python
profile.columns
```

```python
portfolio.columns
```

```python
transcript.columns
```

## --------------------------------------------------------------------------------------

## 4. Removing unwanted columns

```python
profile = profile.drop(columns = ["Unnamed: 0"])
portfolio = portfolio.drop(columns = ["Unnamed: 0"])
transcript = transcript.drop(columns = ["Unnamed: 0"])

print("The column named 'Unnamed :0' has been removed successfully")
```

## ----------------------------------------------------------------------------------

## 5. Verifying Column Removal

```python
print(profile.columns)
print("")
print(portfolio.columns)
print("")
print(transcript.columns)
```

## ----------------------------------------------------------------------------------------

## 6. Inspecting Missing Values

```python
profile[profile["gender"].isnull()].head()
```

```python
profile[profile["income"].isnull()].head()
```

```python
print("Total number of rows which has missing data for column named 'gender' :- ", profile["gender"].isnull().sum())
print("Total number of rows which has missing data for column named 'income' :- ", profile["income"].isnull().sum())
```

## ----------------------------------------------------------------------------------

## 7. Comparing Missing gender and income records

```python
missing_gender = profile[profile["gender"].isnull()]
missing_income = profile[profile["income"].isnull()]

print("Missing Gender shape :- ", missing_gender.shape)
print("Missing income shape :- ", missing_income.shape)
```

```python
missing_gender[["id"]].head()
missing_income[["id"]].head()
```

```python
missing_gender_ids = set(missing_gender["id"])
missing_income_ids = set(missing_income["id"])

print(missing_gender_ids == missing_income_ids)
```

## ---------------------------------------------------------------------------------

## 8. Handling Missing gender values

```python
profile["gender"] = profile["gender"].fillna("Unknown")

print("Missing Gender Values After Cleaning :- ")
print(profile["gender"].isnull().sum())
```

## ---------------------------------------------------------------------------------

## 9. Gender Distribution after cleaning

```python
profile["gender"].value_counts()
```

## ----------------------------------------------------------------------------------

## 10. Investigating income distribution

```python
profile["income"].describe()
```

## ------------------------------------------------------------------------------------

## 11. Handling Missing income values

```python
median_income = profile["income"].median()
print("The median of the income is :- ", median_income)
```

```python
profile["income"] = profile["income"].fillna(median_income)

print("Missing income values after cleaning :- ")
print(profile["income"].isnull().sum())
```

## ---------------------------------------------------------------------------------------

## 12. Verifying Missing values after cleaning

```python
print(profile.isnull().sum())
```

## ---------------------------------------------------------------------------------

## 13. Investigating Age Outliers

```python
profile["age"].describe()
```

```python
profile[profile["age"] == 118].head()
```

```python
profile[profile["age"] == 118].shape
```

```python
profile[profile["age"] == 118][["gender", "age", "income"]].head(10)
```

## ------------------------------------------------------------------------------------

## 14. Handling Age Outliers

```python
profile["age"] = profile["age"].replace(118, np.nan)

print("Missing Values for age :- ")
print(profile["age"].isnull().sum())
```

```python
profile["age"].describe()
```

```python
median_age = profile["age"].median()

print("Median age :- ", median_age)
```

## --------------------------------------------------------------------------------------

## 15. Replacing Missing Age values

```python
profile["age"] = profile["age"].fillna(median_age)

print("Missing age values after cleaning the data :- ")
print(profile["age"].isnull().sum())
```

## -----------------------------------------------------------------------------------

## 16. Final Dataset Verification

```python
print(profile.isnull().sum())
```

## ---------------------------------------------------------------------------------------

## 17. Saving Cleaned Files

```python
profile.to_csv("../cleaned_data/profile_cleaned.csv", index = False)
print("Profile Cleaned Version saved successfully")

portfolio.to_csv("../cleaned_data/portfolio_cleaned.csv", index = False)
print("Portfolio Cleaned Version saved successfully")

transcript.to_csv("../cleaned_data/transcript_cleaned.csv", index = False)
print("Transcript Cleaned Version saved successfully")
```
