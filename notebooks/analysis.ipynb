## 5. Event Distribution

```python
# Which customer events occur most frequently?

master_data["event"].value_counts()
```

### Observation :- 

### Transacation event occur most frequently.  It indicates that purchasing behavior is the most common than promotional interactions.

```python
# What sort of offer types are most common?

master_data["offer_type"].value_counts()
```

```python
master_data_[["age", "income"]].corr()
```

### Observation 

### he correlation between age and income is very weak to moderate. it indicates limited relationship between the two variables.

```python
# which offer types gives the most rewards?

master_data.groupby("offer_type")["reward"].mean()
```

```python
# How difficult are different orders?

master_data.groupby("offer_type")["difficulty"].mean()
```

```python
# Finding average duration by type of order

master_data.groupby("offer_type")["duration"].mean()
```

```python
# Gender Distribution

master_data["gender"].value_counts()
```

```python
# Finding average income based on gender

master_data.groupby("gender")["income"].mean()
```

```python
# Finding average age by gender

master_data.groupby("gender")["age"].mean()
```

```python
# What are the top 10 most used offers?

master_data["offer_id"].value_counts().head(10)
```
