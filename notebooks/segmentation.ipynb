# Customer Segmentation

### This Notebook mainly focuses on grouping customers into different segments according to age and income. The purpose is mainly to understand the customers and their behavior.

## ---------------------------------------------------------------------------------

## 1. Importing Libraries

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

print("Libraries loaded successfully")
```

## -----------------------------------------------------------------------------------

## 2. Loading Master Dataset

```python
master_data = pd.read_csv("../cleaned_data/master_data.csv")

print("Master data loaded successfully")
```

## -------------------------------------------------------------------------------------------

## 3. Select Customer Data

```python
customers = master_data[["person", "gender", "age", "income"]].drop_duplicates()

customers.head()
```

## ---------------------------------------------------------------------------------------

## 4. Create Age Segments

```python
# Creating new column defining them according to their age

def age_segment(age):
    if age < 35:
        return "Young"
    elif age < 60:
        return "Middle Age"
    else :
        return "Senior"

customers["age_segment"] = customers["age"].apply(age_segment)

customers.head()
```

## -------------------------------------------------------------------------------------

## 5. Age Segment Distribution

```python
# Grouping the customers according to their age segment

customers["age_segment"].value_counts()
```

## ----------------------------------------------------------------------------------------

## 6. Create Income Segments

```python
# Creating new column defining them according to their income

def income_segment(income):
    if income < 50000:
        return "Low Income"
    elif income < 80000:
        return "Medium Income"
    else:
        return "High Income"
    
customers["income_segment"] = customers["income"].apply(income_segment)

customers.head()
```

## -------------------------------------------------------------------------------------

## 7. Income Segment Distribution

```python
# Grouping the customers according to their age segment

customers["income_segment"].value_counts()
```

## -------------------------------------------------------------------------

## 8. Average Income by Age Segment

```python
# Customers grouping by age based on their average income

customers.groupby("age_segment")["income"].mean()
```

## ------------------------------------------------------------------------------------

## 9. Average age by income segment

```python
# Customers grouped by income based on their average age

customers.groupby("income_segment")["age"].mean()
```

## ----------------------------------------------------------------------------

## 10. Customer Segment Summary 

```python
customers.groupby(["age_segment", "income_segment"]).size()
```

## ------------------------------------------------------------------------------

## 11. Saving Segmented Customer Data

```python
customers.to_csv("../cleaned_data/customer_segments.csv")

print("Customer segments were saved successfully")
```
