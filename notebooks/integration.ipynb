## --------------------------------------------------------------------------------------

## 8. Extracting Offer IDs from Value Column


```python
import ast 
def extract_offer_id(value):
    value_dict = ast.literal_eval(value)

    if"offer id" in value_dict:
        return value_dict["offer id"]
    elif "offer_id" in value_dict:
        return value_dict ["offer_id"]
    else:
        return None
    
transcript["offer_id"] = transcript["value"].apply(extract_offer_id)
transcript.head()
```

## ----------------------------------------------------------------------------------------

## 9. Verifying extracted offer IDs

##-------------------------------------------------------------------------------------

## 10. Checking Missing offer IDs

```python
print("Total missing offer ids :- ", transcript["offer_id"].isnull().sum())
```

## -------------------------------------------------------------------------------------

## 11. Integrating Transcript and Portfolio

```python
transcript_portfolio = pd.merge(
    transcript,
    portfolio,
    left_on = "offer_id",
    right_on = "id",
    how = "left"
)

print("Shape of the Dataset after merging both transcript and portfolio :- ", transcript_portfolio.shape)
```

```python
transcript_portfolio.head()
```

## ---------------------------------------------------------------------------------------

## 12. Integrating all the data to make master dataset

```python
master_data = pd.merge(
    transcript_portfolio,
    profile,
    left_on = "person",
    right_on = "id",
    how = "left"
)

print("Final Master Dataset Shape :- ", master_data.shape)
```

```python
master_data.head()
```

## ---------------------------------------------------------------------------------------

## 14. Saving Master Data File

```python
master_data.to_csv("../cleaned_data/master_data.csv", index = False)

print("Master dataset saved successfully")
```
